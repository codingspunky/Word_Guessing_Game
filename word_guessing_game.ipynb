```python
def print_game(word):
    print()
    for i in word:
        print(i,end=" ")
```

```python
import random
a = ['pink','yellow','blue','green','black']
b = random.choice(a)

guesses=['_']*len(b)
turns=15
score = 0
temp = ""

print_game(guesses)

while("_" in guesses):
    if(turns == 0):
        break
    
    turns-=1
        
    guess = input("\nenter your alphabet: ")
    
    if(guess in temp):
        print(guess+" is already used")
        continue
    c=False
        

    temp += guess    
        
    for i in range(len(b)):
        if guess==b[i]:
            guesses[i]=b[i]
            c=True
            if guess in 'a,e,i,o,u':
                score+=5
                print("\nyour score is",score)
            else:
                score+=3
                print("\nyour score is",score)
    
    print_game(guesses)
    if not c:
        score-=10
        print("\nyour score is",score)
        
    
print()    
    
    
print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        
if( turns >0):
    print("Score: ",score)
    print("Turns used: ",15-turns)
    print("YOU WON!!ðŸ¥³")
else:
    print("YOU LOST!!ðŸ˜’")
    




```
